```python
import pandas as pd
import numpy as np
import glob
pd.options.display.max_colwidth =1000

# combine all the raw data to one single dataframe
path = r"C:\Users\avci\Documents\MSE\HS20\VT1\Seq2Seq\ami-transcripts"

all_files = glob.glob(path + "/*.txt")

filenames =[]
input_text = []

for i in all_files:
    filenames.append(str(i).split('\\')[-1].split('.transcript.txt')[0])    

for i in all_files:
    f = open(i,'r')
    input_text.append(str(f.read() ))
    f.close()

dfid = pd.DataFrame(filenames, columns=['ID'])
dftxt = pd.DataFrame(input_text, columns=["input_text"])
dfinput = pd.concat([dfid, dftxt], axis=1)
```

```python
# combine all the raw data to one single dataframe
path = r"C:\Users\avci\Documents\MSE\HS20\VT1\Seq2Seq\ami-summary\abstractive"
all_files = glob.glob(path + "/*.txt")

filenames =[]
input_text = []

for i in all_files:
    filenames.append(str(i).split('\\')[-1].split('.abssumm.txt')[0])    

for i in all_files:
    f = open(i,'r')
    input_text.append(str(f.read() ))
    f.close()

dfid = pd.DataFrame(filenames, columns=['ID'])
dftxt = pd.DataFrame(input_text, columns=["target_text"])
dftarget = pd.concat([dfid, dftxt], axis=1)
```

```python
df = pd.merge(dfinput,dftarget, on ="ID")
df = df[['input_text', 'target_text']]
df.to_csv(r'data.txt', header=True, index=False, sep=';', mode='a')
```
